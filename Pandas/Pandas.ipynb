# AGGREGATING FUNCTION

```python
width = [4.1,3.9,2.8,1.4]

data_width = data[data['SepalWidthCm'].isin(width)]
data_width
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
```

```python
print(data['SepalLengthCm'].agg(iqr))
```

```python
print(data[['SepalLengthCm','SepalWidthCm','PetalWidthCm']].agg(iqr))
```

```python
print(data[['SepalLengthCm','SepalWidthCm','PetalWidthCm']].agg([iqr,np.median]))
```
