```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:/study docs/Excel R/Data science/assignment/Multiple linear regression/50_Startups.csv')
data.head()
```

# LINE PLOT

```python
plt.plot(data.index,data['Profit'])
plt.show()
```

```python
plt.plot(data.index,data['Profit'])
plt.plot(data.index,data['Administration'])
plt.show()
```

```python
plt.plot(data.index,data['Profit'],label = "Profit")
plt.plot(data.index,data['Administration'],label = "Administration")
plt.legend()
plt.show()
```

```python
plt.plot(data.index,data['Profit'],label = "Profit")
plt.plot(data.index,data['Administration'],label = "Administration")
plt.title("Marketing Statergies")
plt.ylabel('Profit And R&D')
plt.legend()
plt.show()
```

# STYLING GRAPHS

```python
plt.plot(data.index,data['Profit'],label = "Profit",color = 'DarkCyan')
plt.plot(data.index,data['Administration'],label = "Administration",linestyle = '--')#others are ':' or ''
plt.plot(data.index,data['Marketing Spend'],label = 'Marketing Spend',marker = 'o')#others are 's','d'
plt.title("Marketing Statergies")
plt.ylabel('Profit')
plt.legend()
plt.show()
```

```python
plt.style.use("fivethirtyeight")#others are grayscale,seaborn, classic .....

plt.plot(data.index,data['Profit'],label = "Profit",color = 'DarkCyan')
plt.plot(data.index,data['Administration'],label = "Administration",linestyle = '--')#others are ':' or ''
plt.plot(data.index,data['Marketing Spend'],label = 'Marketing Spend',marker = 'o')#others are 's','d'
plt.title("Marketing Statergies")
plt.ylabel('Profit')
plt.legend()
plt.show()
```

# BAR CHART

```python
plt.bar(data.index,data['Administration'])
plt.xlim(0,50)
```

```python
std_administration = data['Administration'].std()
```

```python
plt.bar(data.index,data['Administration'],yerr = std_administration)
```

```python
plt.bar(data.index,data['Administration'],label = 'Administration')
plt.legend()
plt.xlim(0,50)
```

```python
plt.bar(data.index,data['Administration'],label = 'Administration')
plt.bar(data.index,data['R&D Spend'],label = 'R&D Spend')
plt.xlim(0,50)
plt.legend()
```
